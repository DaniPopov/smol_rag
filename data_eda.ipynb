```python
import json
import pandas as pd 
```

```python
with open('data/stock.json', 'r', encoding='utf-8') as f:
    data = json.load(f)

data[0]
```

```python
for key, value in data[0].items():
    print(key, (value))
```

```python
laptops = 'מחשבים ניידים'
laptops_ids = set()
laptops_categories = set()

for item in data:
    if item['stock_status'] == 'outofstock':
        continue
    for category in item['categories']:
        if laptops == category['name']:
            laptops_ids.add(item['id'])
            break

print(f'Number of laptops ids: {len(laptops_ids)}')
print(laptops_ids)

for item in data:
    if item['id'] in laptops_ids:
        for category in item['categories']:
                laptops_categories.add(category['name'])

print(f'Number of laptops categories: {len(laptops_categories)}')

laptops_categories
```

```python
main_categories = [ 
 'מחשבים ניידים 2in1',
 'מחשבים ניידים Acer',
 'מחשבים ניידים Apple',
 'מחשבים ניידים Asus',
 'מחשבים ניידים Dell',
 'מחשבים ניידים HP',
 'מחשבים ניידים Lenovo',
 'מחשבים ניידים MSI',
]

two_in_one_laptops_ids = set()
acer_laptops_ids = set()
apple_laptops_ids = set()
asus_laptops_ids = set()
dell_laptops_ids = set()
hp_laptops_ids = set()
lenovo_laptops_ids = set()
msi_laptops_ids = set()

for item in data:
    if item['stock_status'] == 'outofstock':
        continue
    if item['id'] in laptops_ids:
        for category in item['categories']:
            if category['name'] in main_categories:
                if category['name'] == 'מחשבים ניידים 2in1':
                    two_in_one_laptops_ids.add(item['id'])
                if category['name'] == 'מחשבים ניידים Acer':
                    acer_laptops_ids.add(item['id'])
                if category['name'] == 'מחשבים ניידים Apple':
                    apple_laptops_ids.add(item['id'])
                if category['name'] == 'מחשבים ניידים Asus':
                    asus_laptops_ids.add(item['id'])
                if category['name'] == 'מחשבים ניידים Dell':
                    dell_laptops_ids.add(item['id'])
                if category['name'] == 'מחשבים ניידים HP':
                    hp_laptops_ids.add(item['id'])
                if category['name'] == 'מחשבים ניידים Lenovo':
                    lenovo_laptops_ids.add(item['id'])
                if category['name'] == 'מחשבים ניידים MSI':
                    msi_laptops_ids.add(item['id'])

# Print statistics
print(f'Number of in-stock laptops ids: {len(laptops_ids)}')
print(f'Number of two in one laptops ids: {len(two_in_one_laptops_ids)}')
print(f'Number of acer laptops ids: {len(acer_laptops_ids)}')
print(f'Number of apple laptops ids: {len(apple_laptops_ids)}')
print(f'Number of asus laptops ids: {len(asus_laptops_ids)}')
print(f'Number of dell laptops ids: {len(dell_laptops_ids)}')
print(f'Number of hp laptops ids: {len(hp_laptops_ids)}')
print(f'Number of lenovo laptops ids: {len(lenovo_laptops_ids)}')
print(f'Number of msi laptops ids: {len(msi_laptops_ids)}')

# Get all laptops that are categorized
all_categorized = set().union(
    two_in_one_laptops_ids,
    acer_laptops_ids,
    apple_laptops_ids,
    asus_laptops_ids,
    dell_laptops_ids,
    hp_laptops_ids,
    lenovo_laptops_ids,
    msi_laptops_ids
)

# Find uncategorized laptops
uncategorized = laptops_ids - all_categorized
print(f'\nUncategorized laptops: {len(uncategorized)}')
```

```python
for item in data:
    if item['id'] in uncategorized:
        print('='*100)
        print(item['name'])
        print(item['id'])
        for category in item['categories']:
            print(category['name'])
        print('='*100)
```
